```python
from PIL import Image
import matplotlib.pyplot as plt

%matplotlib inline
```

# read image with plt.imread

```python
pim = plt.imread('./sample/test.jpg', format='jpeg')
plt.imshow(pim)
print(pim.shape)
print(pim[:,:,0])
```

# read image with Image.open

```python
im = Image.open('./sample/test.jpg')
pix = np.array(im.getdata(), dtype=np.uint8).reshape((951, 634, 3))
```

```python
print(pix[:,:,0])
```

```python
print(np.max(pim-pix))
```

```python
from io import BytesIO

fh_im = BytesIO()
im.save(fh_im, format='JPEG')
fh_im.seek(0)
imgbytes = bytearray(fh_im.read())

# img = skimage.img_as_float(skimage.io.imread(BytesIO(imgbytes), as_grey=False)).astype(np.float32)
img=Image.open(BytesIO(imgbytes))
simg = np.array(img.getdata(), dtype=np.uint8).reshape((951, 634, 3))
print(simg[:,:,0])
```

```python
print(np.max(pim-simg))
```
